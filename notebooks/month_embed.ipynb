```python
import math

import matplotlib.pyplot as plt
import torch
```

```python
x = torch.linspace(0, 2 * math.pi, 100)
plt.plot(x, torch.sin(x))
plt.plot(x, torch.cos(x))
```

```python
months = torch.arange(12)
x = months / 12 * 2 * math.pi
plt.plot(months, torch.sin(x))
plt.plot(months, torch.cos(x))
```

```python
x = (torch.arange(12) / 12) * 2 * math.pi
embeds = torch.vstack((torch.sin(x), torch.cos(x))).T
embeds.shape
```

```python
plt.plot(months, embeds)
```
